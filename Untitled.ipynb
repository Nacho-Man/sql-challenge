```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
#create and engine to start putting this in SQL database
#Note: will not work unless PostGres has a database added with the name EmployeeDB
rds_connection_string = "postgres:postgres@localhost:5432/EmployeeDB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
employee_df = pd.read_sql_query('select * from employees', con=engine)
employee_df
```

```python
salary_df = pd.read_sql_query('select * from salaries', con=engine)
salary_df
```

```python
comb_df = pd.merge(employee_df, salary_df, on='emp_no')
comb_df
```

```python
titles_df = pd.read_sql_query('select * from titles', con=engine)
titles_df
```

```python
comb_df2 = pd.merge(comb_df, titles_df)
comb_df2
```

```python
filt_df = pd.DataFrame(list(zip(comb_df2['title'],comb_df2['salary'])),columns=['Title','Salary'])
filt_df
```

```python
filt_g = filt_df.groupby('Title').mean()
filt_g.head()
```

```python
fig, volume = plt.subplots(figsize= (24, 8))
volume.bar(filt_g.index, filt_g['Salary'])
plt.title('Salary by Title')
fig.savefig('Salary_by_Title.png')

# Show the Figure
fig.show()
```
